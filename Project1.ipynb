```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Calling the Data set
Olympic_file = "olympics.csv"
olym_df = pd.read_csv(Olympic_file)
olym_df
```

```python
# Summary Statistics
olym_df.describe()
```

```python
# Fill NaN with Zeros
oly1_df = olym_df.fillna(0)
oly1_df
```

```python
oly1_df.describe()
```

```python
# Find the top 10 countries with most participating rate 
top_10_countries = oly1_df.Team.value_counts().sort_values(ascending= False).head(10)
top_10_countries
```

```python
# Graphical Representation of Top Ten Countries
plt.figure(figsize = (12,5))
sns.barplot(x= top_10_countries, y= top_10_countries.index)
plt.title("Top ten Participating Countries")
plt.xlabel("Number of Players")
plt.ylabel("Countries")
plt.show()
```

```python
# Graphical representation of Age Distribution of Players
plt.figure(figsize = (12,4))
plt.hist(oly1_df.Age, bins = np.arange(10,60), color = "Blue", edgecolor = "White")
plt.xlabel("Age of Players")
plt.ylabel("Number of Players")
plt.title("Age Distribution of Players")
plt.show()
```

```python
# Names of sports played in winter
winter_games = [oly1_df[oly1_df.Season == "Winter"].Sport.unique()]
winter_games
```

```python
# winter games count
winter_games_count = [oly1_df[oly1_df.Season == "Winter"].Sport.value_counts()]
winter_games_count
```

```python
# Graphical Representation of winter games
plt.figure(figsize = (12,5))
sns.barplot(x= winter_games_count, y= winter_games_count.index)
plt.title("Winter Games")
plt.xlabel("Number of Players")
plt.ylabel("Countries")
plt.show()
```

```python
# Names of Sports played in summer
summer_games = [oly1_df[oly1_df.Season == 'Summer'].Sport.unique()]
summer_games
```

```python
#  Male and Female count  Participant in Olympic History
male_count = oly1_df["Sex"][oly1_df.Sex == "M"].count()
female_count = oly1_df["Sex"][oly1_df.Sex == "F"].count()
total_count = oly1_df.Sex.count()
total_count 
```

```python
# male to female ratio
male_ratio = int((male_count/total_count) * 100)
female_ratio = int((female_count/total_count)*100)
print(f" Male to Female ratio in olympic history is {male_ratio}:{female_ratio}")
```

```python
# Graphical representatio of male to female ratio
users = [male_ratio, female_ratio]
x_axis = np.arange(len(users))
plt.figure(figsize = (6,4))
tick_location = [value for value in x_axis]
plt.xticks(tick_location,["MALE", "FEMALE"])
plt.xlabel("Gender")
plt.ylabel("Number of Players")
plt.title("Ratio of Male to Female Players")
plt.bar(x_axis, users, color = "red", alpha = 0.25, align = "center")
```

```python
# Pie Chart Of Gender Distribution of Olympic Participation
plt.figure(figsize=(10,5))
sex_counts = oly1_df.Sex.value_counts()
plt.title("Male Female Participation ratio in Olympic History")
plt.pie(sex_counts, labels = sex_counts.index, autopct = "%1.1f%%", shadow = True)
plt.show()
```
